```python
import random
random.seed(2015)
import csv
from string import *
```

```python
train_data = open('/Users/junanqu/Desktop/text_p2/train.txt','r').readlines()
test_data =  open('/Users/junanqu/Desktop/text_p2/test.txt','r').readlines()
```

```python
# a customized class to store training-example
class integrated_instance():
    def __init__(self):
        self.sent = ""  
        self.word = []  
        self.pos = [] 
        self.ner = [] 
        self.net_test = [] 
        self.map = {} 
```

```python
def read_data(data, type = "train"):
    if type == "test":
        sent = ""
        for i in range(0, len(data), 3):
            sent += data[i] + "\t"
        pos = ""
        for i in range(1, len(data), 3):
            pos += data[i] + "\t"
        ner = ""
        for i in range(2, len(data), 3):
            ner += data[i] + " "
        
        translation = {None: "\r\n"}
        
        sent = sent.translate(translation)
        pos = pos.translate(translation)
        ner = ner.translate(translation)
        
        test_instance = integrated_instance()
        test_instance.sent = sent[:-1]
        test_instance.word = sent.split("\t")[:-1]
        test_instance.pos = pos.split("\t")[:-1]
        test_instance.ner = ner.split(" ")[:-1]
        return test_instance
    
    list_sent_instance = []    
    
    max_idx_line = len(data)
    idx_line_loaded = 0
    while (idx_line_loaded + 3 <= max_idx_line):
        # create a sent_instance object        
        new_instance = integrated_instance()
        # break strings and         
        new_instance.sent = data[idx_line_loaded][:-2] # -2 to get rid of "\r\n"
        new_instance.word = data[idx_line_loaded][:-2].split("\t")
        new_instance.pos = data[idx_line_loaded+1][:-2].split("\t")
        new_instance.ner = data[idx_line_loaded+2][:-2].split("\t")
        
        list_sent_instance.append(new_instance)
        idx_line_loaded += 3
    return list_sent_instance
```

```python
data = read_data(train_data)
test = read_data(test_data, type="test")
test.word
```

```python
# a function that tranform a list of Sent_instance to a dictionary mapping each word
# to its ner tags. One word can have several ner tags
def word_ner_map(list_instance):
    map_dic = {}     
    for instance in list_instance:
        for i, word in enumerate(instance.word):
#             print(i,word)
            if word not in map_dic.keys():            
                map_dic[word] = []
                map_dic[word].append(instance.ner[i])
            else:
                try:
                    map_dic[word].append(instance.ner[i])
                except IndexError:
                    continue
    return map_dic
```

```python
map_built = word_ner_map(data)
```

```python
def basic_prediction(test_instance, dic_map):
    prediction = {}
    for i in range(len(test_instance.ner)):
        word_to_pred = test_instance.word[i]
        if word_to_pred.find('\n')!=-1:
            word_to_pred = word_to_pred[0:word_to_pred.find('\n')]
        if word_to_pred in dic_map.keys():
            ner_predict = random.choice(dic_map[word_to_pred])
        else:
            ner_predict = 'UNK'
        prediction[i] = (word_to_pred, ner_predict) 
    return prediction
```

```python
# This prediction is pairing every word in test dataset with its NER from traing dataset.

predictions = basic_prediction(test, map_built)

k = 0 
for x in predictions:
    if(predictions[x][1] == 'UNK'):
        k+=1
        print(predictions[x])
print(k)
```

```python
def reformat_for_kaggle(predictions):
    raw_print = {"ORG": [], "MISC": [], "PER": [], "LOC": []}    
    for k, v in predictions.items():
        for ner_tag in raw_print.keys():        
            if ner_tag in v[1]:
                raw_print[ner_tag].append(k)
    
    final_print = {}        
    for key in raw_print.keys():
        l = raw_print[key]
        ranges = sum((list(t) for t in zip(l, l[1:]) if t[0]+1 != t[1]), [])
        iranges = iter(l[0:1] + ranges + l[-1:])
        final_print[key] = [str(n) + '-' + str(next(iranges)) for n in iranges]
    return raw_print, final_print
```

```python
raw_print, final_print = reformat_for_kaggle(predictions)
```

```python
def export(final_print, file_name):
    baseline_submission = [['Type', 'Prediction']]
    for k, v in final_print.items():
        baseline_submission.append([k, " ".join(v)])
    with open(file_name + '.csv', 'w',newline='') as csvfile:
        csv_writer = csv.writer(csvfile, delimiter=',')
        for row in baseline_submission:
            csv_writer.writerow(row)
```

```python
export(final_print, "baseline_submission")
```
